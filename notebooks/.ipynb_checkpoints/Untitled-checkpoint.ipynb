```python
!pip install yfinance

import pandas as pd
import os
import yfinance as yf
from datetime import date, timedelta
import numpy as np
import csv

#Function to fetch EOD data
def fetch_EOD():
    # Create the EOD60 folder if it doesn't exist
    output_folder = "EOD_BackTest"
    os.makedirs(output_folder, exist_ok=True)

    # Read the list of symbols from symbol.csv
    symbol_file = "../EQUITY_L.csv"
    symbols = pd.read_csv(symbol_file)["Symbol"]

    # Set the end date as today
    end_date = date.today()

    # Set the start date as 60 days ago
    start_date = end_date - timedelta(days=700)

    # Fetch historical data for each symbol
    for symbol in symbols:
        try:
            df = yf.download(symbol, start=start_date, end=end_date)
            if not df.empty:
                # Save the data to a CSV file in the EOD folder
                filename = os.path.join(output_folder, f"{symbol}.csv")
                df.to_csv(filename)
                print(f"Data saved for {symbol} in {filename}")
            else:
                print(f"No data available for {symbol}")
        except Exception as e:
            print(f"Error fetching data for {symbol}: {e}")

    print("Data fetching and saving completed.")

 #-------- MAIN PROGRAM STARTS FROM HERE

# Directory containing CSV files
csv_directory = 'EOD_BackTest'
fetch_EOD()
```

```python
import pandas as pd
import numpy as np

def process_data(file_path, output_file_path):
    """
    Processes the CSV file to convert the date format to "DD-MM-YYYY",
    handles infinite or large values, and saves the cleaned data to a new CSV file.

    Args:
        file_path (str): The path to the input CSV file.
        output_file_path (str): The path to save the processed CSV file.
    """
    try:
        # Read the CSV file, parsing dates in the first column
        df = pd.read_csv(file_path, parse_dates=[0])

        # Convert the 'Date' column to datetime objects, handling potential errors
        df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

        # Convert the 'Date' column to the desired "DD-MM-YYYY" format
        df['Date'] = df['Date'].dt.strftime('%d-%m-%Y')

        # Replace infinite values with NaN
        df.replace([np.inf, -np.inf], np.nan, inplace=True)

        # Convert all columns to numeric, coercing errors to NaN
        for col in df.columns[1:]:  # Skip the 'Date' column
            df[col] = pd.to_numeric(df[col], errors='coerce')

        # Drop rows with NaN values
        df.dropna(inplace=True)

        # Save the cleaned DataFrame to a new CSV file
        df.to_csv(output_file_path, index=False)

        print(f"Successfully processed and saved data to {output_file_path}")

    except Exception as e:
        print(f"Error processing file: {e}")

# Example usage:
file_path = 'Book1.csv'  # Replace with your actual file path
output_file_path = 'processed_data.csv'  # Replace with your desired output file path
process_data(file_path, output_file_path)
```
